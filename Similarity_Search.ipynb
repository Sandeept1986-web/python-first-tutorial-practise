```python
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
import torch
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [
    "I love cats",
    "I adore kittens",
    "The stock market crashed"
    ]

#Encode
embeddings = model.encode(sentences)
embeddings.shape
```

```python
cos_sim = cosine_similarity(embeddings[0].reshape(1, -1), embeddings)
print("Cosine similarity with 'I love cats':", cos_sim)

```

```python
dot_sim = np.dot(embeddings[0], embeddings[1])
print("Dot product with 'I love cats':", dot_sim)
```

```python
# Euclidean distance
eucl_dist = euclidean_distances([embeddings[0]], embeddings)
print('Euclidean distance from "I love cats":', eucl_dist)
```

```python
# âœ¨ Try it Yourself!
# Replace the sentences below with your own examples and rerun the notebook.

your_sentences = [
    "Python is a great programming language",
    "I enjoy coding in Python",
    "The weather is nice today"
]

your_embeddings = model.encode(your_sentences)

print("Cosine similarities:")
print(cosine_similarity([your_embeddings[0]], your_embeddings))

print("\nDot product similarities:")
print(np.dot(your_embeddings[0], your_embeddings.T))

print("\nEuclidean distances:")
print(euclidean_distances([your_embeddings[0]], your_embeddings))
```
