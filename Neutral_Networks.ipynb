```python
import numpy as np

class NeutralNetwork:
    def __init__(self, input_size, hidden_size, output_size, lr=0.1):
      self.lr = lr
      self.w1 = np.random.randn(input_size, hidden_size)
      self.w2 = np.random.randn(hidden_size, output_size)
    @staticmethod
    def sigmoid(x):
      return 1 / (1 + np.exp(-x))

    @staticmethod
    def sigmoid_derivative(x):
      return x * (1 - x)

    def forward(self, x):
      self.hidden = self.sigmoid(np.dot(x, self.w1))
      self.output = self.sigmoid(np.dot(self.hidden, self.w2))
      return self.output

    def backward(self, x, y, output):
      error = y - self.output
      d_output = error * self.sigmoid_derivative(self.output)
      d_hidden = (d_output.dot(self.w2.T)) * self.sigmoid_derivative(self.hidden)

      self.w2 += self.hidden.T.dot(d_output) * self.lr
      self.w1 += x.T.dot(d_hidden) * self.lr

    def train(self, x, y, epochs=10000):
      for e in range(epochs):
          output = self.forward(x)
          self.backward(x, y, output)


nn = NeutralNetwork(2, 4, 1)
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])


nn.train(x, y)
print("Final Predictions:")
print(nn.forward(x).round(3))

```
